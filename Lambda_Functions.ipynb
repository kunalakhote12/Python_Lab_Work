```python
#Short code to find the sum of element in list using reduce
from functools import reduce
li=[5,6,7,8,9,10,11]
sum = reduce (lambda x,y :x+y, li)
sum
```

```python
#Short code to find greatest element in list using reduce
from functools import reduce
li=[5,6,7,8,9,10,11]
max = reduce (lambda x,y :x if x>y else y, li)
max
```
